```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Preprocessing
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(nltk.corpus.stopwords.words('english'))

def preprocess_text(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [token for token in tokens if token.isalnum()]
    tokens = [token for token in tokens if token not in stop_words]
    return " ".join(tokens)

processed_faq_data = {preprocess_text(question): answer for question, answer in faq_data.items()}

# Vectorization
tfidf_vectorizer = TfidfVectorizer()
faq_vectors = tfidf_vectorizer.fit_transform(processed_faq_data.keys())

# Chatbot function to respond to user queries
def respond_to_query(query):
    query = preprocess_text(query)
    query_vector = tfidf_vectorizer.transform([query])

    # Find the most similar FAQ question
    similarities = cosine_similarity(query_vector, faq_vectors)
    most_similar_index = np.argmax(similarities)
    response = list(processed_faq_data.values())[most_similar_index]

    return response

# Interactive loop for chatbot
print("Welcome to the Interview Questions Chatbot!")
print("You can ask questions related to interview preparation. Enter 'exit' to end the chat.")

while True:
    user_input = input("You: ")

    if user_input.lower() == 'exit':
        print("Chatbot: Goodbye!")
        break

    response = respond_to_query(user_input)
    print("Chatbot:", response)
```
